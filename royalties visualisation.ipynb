```python
import glob
```

```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
artists_n_splits = [{
        "artist_name": "TIM-Z",
        "split": 75
    },{
         "artist_name": "RealDJTroy",
         "split": 75
    },{
         "artist_name": "Luchandry Damons",
         "split": 75
    },{
         "artist_name": "Blour",
         "split": 75
    },{
         "artist_name": "Micsy Mohr",
         "split": 75
    },{
         "artist_name": "Bulinki",
         "split": 75
    },{
         "artist_name": "Foza De MC",
         "split": 75
    },{
         "artist_name": "DJ Obza",
         "split": 75
    },{
         "artist_name": "Xplissit",
         "split": 60
    }
]

distribution = {
    "release_info": {
        "artist": [],
        "streams": [],
        "aritst_revenue": [],
        "ba_revenue": [],
        "total_revenue": []
    },
    "total_streams": 0,
    "total_revenue_for_artists": 0,
    "total_revenue_for_ba": 0
}

total_streams = 0
total_revenue_for_artists = 0
total_revenue_for_ba = 0
```

```python
all_files = glob.glob('./csv/*.csv')
df_from_each_file = (pd.read_csv(f) for f in all_files)
royalties_data = pd.concat(df_from_each_file)
royalties_data.columns = royalties_data.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
```

```python
for splits in artists_n_splits:
    artist_data_frame = royalties_data[royalties_data.artist.str.contains(splits["artist_name"])]
    artist_total = artist_data_frame.amount_due_in_usd.sum()
    artist_streams = artist_data_frame.quantity.sum()
    artist_split = (splits["split"]/100) * artist_total
    black_atmosphere_split = artist_total - artist_split
    
    total_streams += artist_streams
    total_revenue_for_artists += artist_split
    total_revenue_for_ba += black_atmosphere_split
    
    artist_total = format(artist_total, '.2f')
    artist_split = format(artist_split, '.2f')
    black_atmosphere_split = format(black_atmosphere_split, '.2f')
    
    distribution["release_info"]["artist"].append(splits["artist_name"])
    distribution["release_info"]["streams"].append(artist_streams)
    distribution["release_info"]["aritst_revenue"].append(artist_split)
    distribution["release_info"]["ba_revenue"].append(black_atmosphere_split)
    distribution["release_info"]["total_revenue"].append(artist_total)

    print("=============================")
    print("Artist Name: " + str(splits["artist_name"]))
    print("Artist Total Revenue: " + str(artist_total))
    print("Artist Total Streams: " + str(artist_streams))
    print("Artist Split: " + str(artist_split))
    print("Black Atmosphere Split: " + str(black_atmosphere_split))

distribution["total_streams"] = total_streams
distribution["total_revenue_for_artists"] = total_revenue_for_artists
distribution["total_revenue_for_ba"] = total_revenue_for_ba

print("=============================")
print("Total Streams: " + str(total_streams))
print("\n*****************************")
print(str(distribution))
```
